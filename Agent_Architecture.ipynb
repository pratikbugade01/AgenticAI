### 1) Reactive Agent

```python
def reactive_agent(environment):
    if environment == "obstacle":
        return "Bhago..!!"
    elif environment == "target":
        return "Chalo chalte hain..!!"
    else:
        return "Ruko jara..!!"
    
    
```

```python
print(reactive_agent("obstacle"))
```

```python
print(reactive_agent("target"))
```

```python
print(reactive_agent("gd"))
```

### 2) Model based Agent

```python
class ModelbasedAgent:
  def __init__(self):
    self.state = None

  def act(self,perception):
    self.state = perception
    if self.state == "danger":
      return "Bhago..!"
    return "Chalthe raho.!"
```

```python
agent = ModelbasedAgent()
print(agent.act("danger"))
```

### 3) Goal Based Agent

```python
def goalbasedagent(currentstate,goalstate):
    if currentstate != goalstate:
        return "take the actions towards the goal"
    return "goal achieved...!"

print(goalbasedagent("M","Z"))
```

### 4) Utility Based Agent

```python
actions = {
    "path_A" : 50,
    "path_B" : 100,
    "path_c" : 30
}  

bestaction = max(actions,key = actions.get)
print("Best action:", bestaction)
```
