### Tokanization

```python
from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("gpt2")

text = "LLMs are powerful"
tokens = tokenizer.tokenize(text)
token_ids = tokenizer.encode(text)

print("Tokens:",tokens)
print("Token IDs:", token_ids)
```

### Token Prediction Sample code

```python
import random

vocab = ["AI", "will", "change", "the", "world"]

def next_token_prediction(prompt):
    return random.choice(vocab)

prompt = "AI will"
for i in range(5):
    next_word = next_token_prediction(prompt)
    prompt += " " + next_word

print(prompt)
```

```python
from transformers import GPT2LMHeadModel
import torch
model = GPT2LMHeadModel.from_pretrained("gpt2", output_attentions=True)
inputs = tokenizer("AI will change the future", return_tensors="pt")

with torch.no_grad():
    outputs = model(**inputs)

attention = outputs.attentions[0]
print(attention.shape)
```
